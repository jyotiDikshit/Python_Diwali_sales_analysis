# Exploratory Data Analysis

```python
#visualizing data
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df.columns
```

# Gender

```python
a=sns.countplot(x='Gender' ,data=df)
for bars in a.containers:
    a.bar_label(bars)
```

```python
df.groupby(['Gender'])['Amount'].sum()
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum()

sns.barplot(x='Gender',y='Amount',data= sales_gen)
```

from the above 2 graphs we can see that most of th buyers are females and even purchasing power of female are greater than male.

# Age

```python
a=sns.countplot(x='Age Group' ,data=df,hue='Gender') #hue basically use to see on the basis of division i.e. male or female division
for bars in a.containers:
    a.bar_label(bars)
```

```python
#total amount vs to
sales_gen=df.groupby(['Age Group'],as_index=False)['Amount'].sum()
sns.barplot(x='Age Group',y='Amount',data= sales_gen)
```

from the above graphs we can see that most of the buyer are of age group 26-35 yrs female

# state

```python
#total number of orders from top 10 states
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
```

```python
sns.set({'figure.figsize':(15,5)})
sns.barplot(x='State',y='Orders',data=sales_state)
```

# total amount/sales from top 10 state

```python
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set({'figure.figsize':(15,5)})
sns.barplot(x='State',y='Amount',data=sales_state)
```

from the above graphs we can see that most of the  orders and total sales  are from U.P.,Maharashtra,karnataka respectively.

# Marital status

```python
a=sns.countplot(x='Marital_Status' ,data=df)
for bars in a.containers:
    a.bar_label(bars)
```

```python
sales_state=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set({'figure.figsize':(10,5)})
sns.barplot(x='Marital_Status',y='Amount',data=sales_state,hue='Gender')
```

from the above graph we can see that most of buyers are married (women) and purchasing power also high of them.

# occupation

```python
#to see no.of orders ob the basis of occupation
sns.set({'figure.figsize':(20,5)})
a=sns.countplot(data=df,x='Occupation')
for bars in a.containers:
    a.bar_label(bars)
```

```python
#purchasing power of different sector
sales_state=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set({'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='Occupation',y='Amount')
```

from the above graph we can see that buyers maximum  in  IT sector,Healthcare,Aviation.

# Product category

```python
sns.set({'figure.figsize':(30,10)})
a=sns.countplot(data=df,x='Product_Category')
for bars in a.containers:
    a.bar_label(bars)
```

```python
sales_state=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set({'figure.figsize':(20,8)})
sns.barplot(data=sales_state,x='Product_Category',y='Amount')
```

from the above graph we can see most of the sold  products are from food,clothing & electronics category.

# conclusion:
    

Married women age group 26-35 yrs from UP,Maharashtra & Karnataka working in  IT,Healthcare,&aviation are more likely to buy  products from food,clothing,electronics category.
